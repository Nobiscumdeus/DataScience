### Dealing with missing data 

```python
#### Identify missing values
print("Not a number",cas.isna().sum())
print("Null: ",cas.isnull().sum())
```

### Identify missing values 

```python
missing=cas[cas.isnull().any(axis=1)]
missing
```

#### Common Approach to missing Values 
##### 1) Fill the missing values by the mean/median, in case of numerical variable 
##### 2) Fill the missing values with the class which has maximum count, in case of categorical variable 

```python
cas.describe()
```

```python
cas['Score'].mean()
```

```python
cas['Score'].fillna(cas['Score'].mean(),inplace=True)
```

```python
#### Calculating the median
cas['Score'].median()
```

```python
cas['Score'].value_counts()
```

```python
cas['Score'].value_counts().index[0]
```

```python
cas['Score'].mode()
```

```python
##You can fill NA\NaN values using the specified value
cas['Score'].fillna(cas['Score'].mode()[0],inplace=True)
#and then check for the total null values again 
cas.isnull().sum()
```

```python
##To fill NA\NaN values in both numerical and categorical variables at one stretch 
cas=cas.apply(lambda x:x.fillna(x.mean()) if x.dtype=='float' else\
    x.fillna(x.value_counts().index[0]))
```
