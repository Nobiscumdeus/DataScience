### Members of Group 38 and what they filled their Group as in the Google form

```python
Group_38=data.loc[data['Group No:']=='38']
Group_38
weird=data.loc[data['Group No:']=='No 38']
weird
another_weird=data.loc[data['Group No:']=='Group 38']
another_weird
display('Total of all memebers ')
Group_38.count()+weird.count()+another_weird.count()
display('Concatenation of all Group 38 members ')
full_group_38=display(pd.concat([Group_38,weird,another_weird],keys=['Used 38 as Group','Used No 38 as Group','Used Group 38 as Group']))
```

### Scores of Students in GROUP 38 in descending order

```python
all_group_38=pd.concat([Group_38,weird,another_weird])
all_group_38.sort_values(by='Score',ascending=False,kind='mergesort')
all_group_38.reset_index(drop=True)
```

### Total number of students in GROUP 38

```python
all_group_38.count()

print(f"There are {all_group_38.shape[0]} students in GROUP 38")
```
