#### Data Visualization Library

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Tips Daaset from Seaborn
tips=sns.load_dataset('tips')
file_path='../Data/tips.csv'
tips.to_csv(file_path,index=False)
tips_data=pd.read_csv('./tips.csv')
```

```python
tips_data.head(n=5)
```

```python
sns.relplot(data=tips_data,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
iris_data=sns.load_dataset('iris')
iris_data.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris_data)
```

```python
titanic=sns.load_dataset('titanic')
titanic.to_csv('../Data/titanic.csv',index=False)
titanic_data=pd.read_csv('../Data/titanic.csv')
titanic_data.head()
```

```python
sns.countplot(x='class',data=titanic_data)
```

```python
sns.countplot(x='survived',data=titanic_data)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
