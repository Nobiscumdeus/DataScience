### MatplotLib Package 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

##### Plotting the Data 

```python
plt.plot(x,y)
plt.xlabel('Values of X (random values between 0 and 10')
plt.ylabel('Sine of random values of x ')
plt.title('Sine Curve')
plt.show()
```

```python
### Adding decorations
plt.plot(x,z)
plt.xlabel('Values of X ')
plt.ylabel('Cosine Values of X')
plt.title('Cosine Curve ')
plt.show()
```

```python
a=np.linspace(-10,20,40)
b=x**2
plt.plot(x,y,'r+')
plt.show()
```

```python
a=np.linspace(-10,10,30)
b=a**2
plt.plot(a,b,'bo')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('People')

plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.xlabel('Languages')
plt.ylabel('People')

plt.show()
```

#### Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
d=np.log(x)
e=np.tan(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
ax.scatter(x,e,color='y')
plt.xlabel('X  values ')
plt.ylabel('Tangent Values')
plt.show()
```

```python
### 3d plot
fig3=plt.figure()
ax=plt.axes(projection='3d')
m=20*np.random.random(100)
n=np.sin(m)
o=np.cos(m)
plt.scatter(n,o,m,c=n,cmap='Blues')
plt.show()
```
